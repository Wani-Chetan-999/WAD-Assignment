```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('air-quality-india.csv')
df = data.iloc[::10]  # Now it's ~2000 rows
df
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Timestamp'], df['PM2.5'], color='blue')
plt.title('PM2.5 Levels Over Time')
plt.xlabel('Timestamp')
plt.ylabel('PM2.5')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='PM2.5', data=df, estimator='mean', palette='viridis')
plt.title('Average PM2.5 by Month')
plt.xlabel('Month')
plt.ylabel('Average PM2.5')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['PM2.5', 'Year', 'Month', 'Day', 'Hour']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Hour', y='PM2.5', alpha=0.6)
plt.title('PM2.5 vs Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('PM2.5')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['PM2.5'], kde=True, bins=30, color='teal')
plt.title('PM2.5 Distribution')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Hour', y='PM2.5', data=df)
plt.title('Boxplot of PM2.5 by Hour')
plt.xlabel('Hour')
plt.ylabel('PM2.5')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Month', y='PM2.5', data=df, inner='quartile', palette='Set2')
plt.title('Violin Plot of PM2.5 by Month')
plt.xlabel('Month')
plt.ylabel('PM2.5')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Timestamp', y='PM2.5', data=df)
plt.title('Time Series of PM2.5')
plt.xlabel('Timestamp')
plt.ylabel('PM2.5')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
